```
path_train_fire='/content/drive/My Drive/dataset/training_dataset/fire'
files_train_fire = os.listdir(path_train_fire)

train_fire_label=[]
train_fire_path=[]
for dir in files_train_fire:
  train_fire_label.append(dir)
  img_path=os.path.join(path_train_fire,dir)
  train_fire_path.append(img_path)


batch_size=64



count=0
for (b,i) in enumerate(range(0,len(train_fire_path),batch_size)):
  print(count)
  batch_images=[]
  batch_path=train_fire_path[i:i+batch_size]
  batch_label=train_fire_label[i:i+batch_size]
  if(count>=48):
    for img_path in batch_path:
      print(img_path)
      image = load_img(img_path, target_size=(224, 224))
      image = img_to_array(image)
      image = np.expand_dims(image, axis=0)
      image = imagenet_utils.preprocess_input(image)
      batch_images.append(image)
    batch_images=np.vstack(batch_images)
    inc_model = InceptionV3(weights="imagenet",include_top=False)
    features=inc_model.predict(batch_images,batch_size=batch_size)
    features = features.reshape((features.shape[0], 5*5*2048))
    for (label, vec) in zip(batch_label, features):
      vec = ",".join([str(v) for v in list(vec)])
      with open(r'/content/drive/My Drive/updated_train.csv', 'a') as f:
        f.write("{},{},{}\n".format(label, vec,'1'))
  count+=1




```

EXTRACTING X FROM CSV

```
chunksize = 1
for chunk in pd.read_csv('/content/drive/My Drive/updated_train.csv', chunksize=chunksize):
    X=chunk.iloc[:,:].values
    break

chunksize=64
count=0
for chunk in pd.read_csv('/content/drive/My Drive/updated_train.csv', chunksize=chunksize):

  if(count<25 or count>60):

      k=chunk.iloc[:,:].values
      X=np.append(X,k,axis=0)

  print(count)
  count+=1
  print(X.shape)



```

```
y=X[1:,-1]
X=X[1:,1:-1]
```

```
print(X.shape,y.shape)
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 57)

```

```
import pickle
with open('/content/drive/My Drive/train_final.pkl','wb') as f:
  pickle.dump((X,y),f)
```

```
print(X_train)
```

```
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X=sc.transform(X)
```

```
y_train=y_train.astype(int)
y_test=y_test.astype(int)
```

```
def print_conf(y_test,y_pred,method):
  cm = confusion_matrix(y_test, y_pred)
  print(method)
  print(cm)
  print ('Accuracy Score :',accuracy_score(y_test, y_pred))

```

```
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)
print_conf(y_test,y_pred,'logistic regression')
```

```
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)
print_conf(y_test,y_pred,'knn-5')
```

```
classifier = KNeighborsClassifier(n_neighbors = 3, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)
print_conf(y_test,y_pred,'knn-3')
```

```
classifier = SVC(kernel = 'rbf', random_state = 0)
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)
print_conf(y_test,y_pred,'SVM')
```

```
classifier = GaussianNB()
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)
print_conf(y_test,y_pred,'Naive Bayes')
```

```
classifier = DecisionTreeClassifier(criterion = 'gini', random_state = 50)
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)
print_conf(y_test,y_pred,'DecisionTree')
```

```
classifier = RandomForestClassifier(n_estimators = 128, criterion = 'gini', random_state = 50)
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)
print_conf(y_test,y_pred,'RandomForest')
```
